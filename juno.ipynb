```python
import cv2
```

```python
# Load pre-trained HOG people detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
# Load input image
image_path = 'image.jpg'
image = cv2.imread(image_path)
```

```python
# Perform human body detection
bodies, _ = hog.detectMultiScale(image, winStride=(8, 8), padding=(16, 16), scale=1.05)
```

```python
# Draw rectangles around detected bodies
for (x, y, w, h) in bodies:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
# Display the output image
cv2.imshow('Human Body Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
