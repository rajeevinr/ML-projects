```python
# ==============================
# STEP 1: IMPORT LIBRARIES
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# ==============================
# STEP 2: LOAD DATA
# ==============================
df = pd.read_excel("responses.xlsx")

# ==============================
# STEP 3: DATA CLEANING
# ==============================
# Remove unwanted columns
df = df.drop(columns=['Timestamp'], errors='ignore')

# Convert Likert scale to numbers
mapping = {
    "Strongly Agree": 5,
    "Agree": 4,
    "Neutral": 3,
    "Disagree": 2,
    "Strongly Disagree": 1
}

df = df.replace(mapping)

# Keep only numeric columns
df = df.select_dtypes(include='number')

# ==============================
# STEP 4: DESCRIPTIVE STATISTICS
# ==============================
print("\nDescriptive Statistics:")
print(df.describe())

# ==============================
# STEP 5: CATEGORY CLASSIFICATION
# ==============================

efficiency = df.iloc[:, 0:3]
responsiveness = df.iloc[:, 3:6]
personalization = df.iloc[:, 6:9]
problem_solving = df.iloc[:, 9:11]
trust = df.iloc[:, 11:14]
satisfaction = df.iloc[:, 14:17]
reuse = df.iloc[:, 17:19]

# Category means
category_scores = pd.DataFrame({
    "Efficiency": efficiency.mean(axis=1),
    "Responsiveness": responsiveness.mean(axis=1),
    "Personalization": personalization.mean(axis=1),
    "Problem_Solving": problem_solving.mean(axis=1),
    "Trust": trust.mean(axis=1),
    "Satisfaction": satisfaction.mean(axis=1),
    "Reuse": reuse.mean(axis=1)
})

print("\nCategory Means:")
print(category_scores.mean())

# ==============================
# STEP 6: VISUALIZATION
# ==============================

# Bar Chart
category_scores.mean().plot(kind='bar')
plt.title("Category-wise Average Scores")
plt.xticks(rotation=45)
plt.show()

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(category_scores.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()

# ==============================
# STEP 7: REGRESSION ANALYSIS
# ==============================

X = category_scores[['Efficiency','Responsiveness','Personalization','Problem_Solving','Trust']]
y = category_scores['Satisfaction']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())

# ==============================
# STEP 8: SAVE OUTPUT
# ==============================
category_scores.to_csv("final_analysis.csv", index=False)

print("\n✅ Analysis Complete")
```
